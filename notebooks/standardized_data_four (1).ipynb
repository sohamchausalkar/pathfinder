```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
data = pd.read_csv("standardized_data_four - standardized_data_three.csv (1).csv")
```

```python
data
```

```python
data.dtypes
```

```python
data.fillna(0, inplace=True)
```

```python
data
```

```python
data.isna()
```

```python
encoder = OneHotEncoder()
```

```python
data.dtypes
```

```python
transformer = ColumnTransformer(
    transformers=[
        ("oneHotEncoder", encoder, ['Program_Name', 'Organization_Name', 'Description_of_Program',
                                      'Link', 'Type', 'Educationa_Level', 'Student_criteria', 'Field_of_Study', 'Skills_preferred', 'Location', 'Timezone', 'Mode_of_Participation', 'Program_Duration', 'Deadline', 'Tuition_or_Fees', 'Stipend', 'Keywords', 'Tags']),
    ],
    remainder="passthrough"
)
```

```python
data.columns = data.columns.str.strip()
```

```python
print(data.columns.tolist())
```

```python
X_transformed = transformer.fit_transform(data)
```
